### **Exercise 4 â€” Comparing Modules**



1. Compute the **mean** and **standard deviation** of attendance per module using pandas `groupby`. Consult the documentation if needed.


````python
# Mean and standard deviation are of the attendance %s per class, rather than total number of attendees
stats = df.groupby("Module Name")["Attended"].agg(["mean", "std"]).reset_index()

stats # Output
````

2. Create a `seaborn.barplot` of mean attendance per module with 95% confidence intervals.


````python
import seaborn as sns

plt.figure(figsize = (20,10))

# Creates a barplot w/ seaborn from the dataframe
sns.barplot(
    data = df,
    x = "Module Name",
    y = "Attended",
    estimator = "mean",
    errorbar = ('ci', 95) # 95% confidence interval
)

plt.title("Mean Attendance Per Module")
plt.xticks(rotation = 45, ha = "right") # Rotates the text on the x axis to fit
plt.xlabel("Module Name", labelpad = 30)
plt.ylabel("Average Attendance", labelpad = 30)
plt.grid(True)
plt.show()
````

### **Exercise 5 â€” Attendance Relationships**



1. Answer the question: **Do students with low overall attendance tend to attend modules that have low attendance?**.

2. **[For L5 students only]**. Choose between a correlation analysis or a linear regression analysis to answer the question.

3. **[For L6 students only]** Run both correlation analysis and linear regression analysis to answer the same question. Do they tell the same story?

> Tips:
> 1. Compute the mean attendance rate of each module. Can you re-use some of the `groupby` code you wrote above?
> 2. For each module, compute the average overall attendance of the students who attended that module.
> 3. Correlation and linear regression can both help you answer the question. 
> 4. Report main statistic and p-value(s).


````python
from scipy.stats import pearsonr

# The attendance rate of each student, sorted by their unique Person Code
rate_per_student = (
    df.groupby("Person Code")["Attended"]
      .mean()
      .reset_index()
      .rename(columns = {"Attended": "Overall Attendance"})
)

# rate_per_student.head()

# The average attendance across all classes of the students who attended that module
mod_rate = (
    df.merge(rate_per_student, on = "Person Code") # Uses Person Code as the unique key to merge dataframes
      .groupby("Module Name")["Overall Attendance"]
      .mean()
      .reset_index()
)

# mod_rate.head()

# Find correlation and p-value
corr, p_value = pearsonr(stats["mean"], mod_rate["Overall Attendance"])
print("Correlation:", corr)
print("P-value:", p_value)
````
